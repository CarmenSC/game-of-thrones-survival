```python
import pandas as pd
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../features_for_quiz/houses/classes.csv')
```

```python
data
```

```python
X = data.drop(columns='class')
y = data['class']
```

```python
tsne = TSNE(perplexity=10, init='random', n_iter=1_000_000)
X_embedded = pd.DataFrame(tsne.fit_transform(X), columns=tsne.get_feature_names_out())
```

```python
X_embedded
```

```python
colors = ['red', 'blue', 'green', 'orange', 'purple']
markers = ['o', 'x', '*', '2']
```

```python
plt.figure()
c = 0
s = 0
for i in range(len(X_embedded)):
    plt.scatter(X_embedded.loc[i][0], X_embedded.loc[i][1], c=colors[c%5], marker=markers[s%4], label=y[i])
    c += 1
    s += 1
plt.legend(bbox_to_anchor=(1, 1.1))
plt.title('t-SNE visualization')
plt.xlabel('tsne0')
plt.ylabel('tsne1')
plt.show()
```

```python
plt.figure()
sns.scatterplot(data = pd.DataFrame(X_embedded), x = 'tsne0', y = 'tsne1', hue=y);
plt.legend(bbox_to_anchor=(1.1, 1.05));
```

```python
import pandas as pd
import re

df_char = pd.read_csv('../raw_data/scrapped_characters.csv')
df_house = pd.read_csv('../raw_data/scrapped_houses.csv')
```

```python
df_char['Character_Number'] = df_char['URL'].str.extract(r'(\d+)$')
```

```python
example = df_char[df_char['Name'] == 'Addam Velaryon']['Allegiances'].values
print(example)
```

```python
df_char['Allegiances_Number'] = df_char['Allegiances'].apply(lambda x: re.findall(r'\d+', x))
```

```python
# checking for one example to see if it works
print(df_char[df_char['Name'] == 'Addam Velaryon']['Allegiances_Number'].values)
```

```python
print(df_char.columns)
df_char.head(3)
```

```python
df_house['Allegiances_Number'] = df_char['URL'].str.extract(r'(\d+)$')
df_house.head()
```

```python
# checking the max of allegiances
max_allegiances = df_char['Allegiances_Number'].apply(len).max()
print("Maximum number of allegiance numbers a person can have:", max_allegiances)


person_with_max_allegiances = df_char.loc[df_char['Allegiances_Number'].apply(len).idxmax(), 'Name']
print("Person with the maximum number of allegiances:", person_with_max_allegiances)


# show the person with the maximum number of allegiances and their allegiances
df_char[df_char['Name'] == person_with_max_allegiances]['Allegiances_Number'].values

num_characters_with_multiple_allegiances = len(df_char[df_char['Allegiances_Number'].apply(len) > 1])
print("Number of characters with more than one allegiance:", num_characters_with_multiple_allegiances)
```

```python
characters_with_multiple_allegiances = df_char[df_char['Allegiances_Number'].apply(len) > 1]
sorted_characters = characters_with_multiple_allegiances.sort_values(by='Allegiances_Number', ascending=False)
sorted_characters.head(5)
```

```python
# Get the maximum number of allegiances
max_allegiances = df_char['Allegiances_Number'].apply(len).max()

# Split the allegiances values into separate columns
for i in range(1, max_allegiances + 1):
    column_name = f"Allegiances_Number_{i}"
    df_char[column_name] = df_char['Allegiances_Number'].apply(lambda x: x[i-1] if len(x) >= i else None)
```

```python
df_char.head(2)
```

```python
# look in df_house for names of houses using Allegiances_Number and add them to allegiances_number_1, allegiances_number_2, etc. in the df_char

# create a dictionary with the house names and their numbers
house_dict = dict(zip(df_house['Allegiances_Number'], df_house['Name']))
house_dict

# replace the numbers in the df_char with the house names
for i in range(1, max_allegiances + 1):
    column_name = f"Allegiances_Number_{i}"
    df_char[column_name] = df_char[column_name].replace(house_dict)

# check if it worked for one example with max allegiances
df_char[df_char['Name'] == person_with_max_allegiances][['Name', 'Allegiances_Number_1', 'Allegiances_Number_2', 'Allegiances_Number_3', 'Allegiances_Number_4']]
```

```python
allegiances_columns = ['Allegiances_Number_1', 'Allegiances_Number_2', 'Allegiances_Number_3', 'Allegiances_Number_4']
house_counts = df_char[allegiances_columns].stack().value_counts()
print(house_counts)
```

```python
# Get the total count of each allegiance number
allegiance_counts = df_char[allegiances_columns].apply(pd.Series.value_counts).sum()
allegiance_counts
```

```python
# get the most common allegiance for the characters
df_char['Most_Common_Allegiance'] = df_char[['Allegiances_Number_1', 'Allegiances_Number_2', 'Allegiances_Number_3', 'Allegiances_Number_4']].mode(axis=1)[0]
```

```python
characters_with_multiple_allegiances = df_char[df_char['Allegiances_Number'].apply(len) > 1]
selected_characters = characters_with_multiple_allegiances[['Name', 'Allegiances_Number_1', 'Allegiances_Number_2', 'Allegiances_Number_3', 'Allegiances_Number_4','Most_Common_Allegiance']]
selected_characters
```

```python
df_char['Allegiance_Count'] = df_char['Allegiances_Number_1'].map(house_counts) + df_char['Allegiances_Number_2'].map(house_counts) + df_char['Allegiances_Number_3'].map(house_counts) + df_char['Allegiances_Number_4'].map(house_counts)
```

```python
df_char.drop(columns=['Allegiance_Count'], inplace=True)
```

```python
# check how many regions there are in the house DataFrame
num_regions = df_house['Region'].nunique()
print("Number of different regions in the house DataFrame:", num_regions)
```

```python
# check how many houses are in each region
house_counts_per_region = df_house['Region'].value_counts()
print(house_counts_per_region)
```

```python
df_char.head()
```

```python
df_char['First_Allegiance_Number'] = df_char['Allegiances_Number'].apply(lambda x: x[0] if x else None)
```

```python
# check if it worked for one example with max allegiances
df_char[df_char['Name'] == person_with_max_allegiances][['Name', 'Allegiances_Number','Allegiances_Number_1', 'Allegiances_Number_2', 'Allegiances_Number_3', 'Allegiances_Number_4','First_Allegiance_Number']]
```

```python
df_char.head()
```

```python
merged_df = df_char.merge(df_house, left_on='First_Allegiance_Number', right_on='Allegiances_Number', how='left')
```

```python
merged_df.columns
```

```python
merged_df.rename(columns={'Name_x': 'Character_Name',
                          'Titles_x': 'Character_Titles',
                            'Name_y': 'House_Allegiance',
                            'Region': 'House_Region',
                            'Title_y': 'House_Titles'
                          }, inplace=True)
```

```python
merged_df.drop(columns=['Allegiances_Number_x',
                        'Allegiances_Number_1',
                        'Allegiances_Number_2',
                        'Allegiances_Number_3',
                        'Allegiances_Number_4',
                        'URL_x',
                        'URL_y',
                        'Allegiances_Number_y'], inplace=True)
```

```python
merged_df.columns
```

```python
# extract number from overloard
merged_df['Overloard_Number'] = merged_df['Overlord'].str.extract(r'(\d+)$')
```

```python
df_house[['URL','Allegiances_Number']].head(2)
```

```python
df_merged = merged_df.merge(df_house[['Allegiances_Number', 'Name']], left_on='Overloard_Number', right_on='Allegiances_Number', how='left')
df_merged.rename(columns={'Name': 'Overlord_Name'}, inplace=True)
df_merged.drop(columns=['Allegiances_Number'], inplace=True)
```

```python
df_merged.columns
```

```python
df_merged.head(2)
```

```python
df_merged['Father'] = df_merged['Father'].apply(lambda x: 1 if pd.notnull(x) else 0)
df_merged['Mother'] = df_merged['Mother'].apply(lambda x: 1 if pd.notnull(x) else 0)
df_merged['Spouse'] = df_merged['Spouse'].apply(lambda x: 1 if pd.notnull(x) else 0)
```

```python
from diffusers import AutoPipelineForText2Image
import torch

pipeline = AutoPipelineForText2Image.from_pretrained(
    "runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16, use_safetensors=True
).to("cuda")
```

```python
prompt = "I'm gonna give you some information on a made up character in the world of game of thrones, please create a short funny story about their death, age is 26, member of house stark"
pipeline(prompt, num_inference_steps=30).images[0]
```
